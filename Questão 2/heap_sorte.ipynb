```python
import matplotlib.pyplot as plt
import numpy as np
import time
import timeit
import math
```

```python
def heapify(arr, n, i): 
    largest = i 
    l = 2 * i + 1     
    r = 2 * i + 2     
  
    if l < n and arr[i] < arr[l]: 
        largest = l 
  
    if r < n and arr[largest] < arr[r]: 
        largest = r 
  

    if largest != i: 
        arr[i],arr[largest] = arr[largest],arr[i] # swap 
  
        heapify(arr, n, largest) 
  
def heapSort(arr): 
    n = len(arr) 
    # maxheap. 
    for i in range(n, -1, -1): 
        heapify(arr, n, i) 
  
    for i in range(n-1, 0, -1): 
        arr[i], arr[0] = arr[0], arr[i] # swap 
        heapify(arr, i, 0)
```

```python
def gera_valor_aleatorio(n):
    arr=(np.random.randint(0,1000,size=n))
    return  (arr)
def gera_valor_ordenado(n):
    return(np.arange(n))
def gera_valor_ordenado_descrecente(n):
    return(np.arange(n,0,-1))
```

O tempo de execução do ''PENEIRA''  é Ø(1) para corrigir o, mais o tempo que o algoritmo executa recursivamente se comporta como uma arvore binaria.As sub-arvores de cada nó têm tamanho maximo de 2n=3  o pior caso acontece quando o último nível da árvore está exatamente metade cheio é o tempo de execução pode ser expresso pela recorrência:
T(n) <= T(2n=3) + Ø(1).
se resolvemos essa recorrencia vamos achar o tempo de execução do algoritmo peneira é O(log2(n)) e o algoritmo
max_heap pode se limitado por O(n),como o maior peso do algoritmo heap sort está no procedimento max_heap e peneira 
seu O(max_heap*peneira)=O(nlog2(n))


```python

arr=[]
arr1=[]
arr2=[]
tempo=[]
tempo1=[]
tempo2=[]
intervalo=1000

for n in range(0,10000,intervalo):#gerando valores aleatorios
    vet=gera_valor_aleatorio(n)      
    inicio = timeit.default_timer()     
    heapSort(vet)
    arr.append(vet)
    fim = timeit.default_timer()
    tempo.append(fim-inicio)
    
for n in range(0,10000,intervalo):#gerando valores ordenados ordem crescente
    vet1=gera_valor_ordenado(n)
    inicio1 = timeit.default_timer()
    heapSort(vet1)
    arr1.append(vet1)
    fim1 = timeit.default_timer()
    tempo1.append(fim1-inicio1)

    
for n in range(0,10000,intervalo):
    vet2=gera_valor_ordenado_descrecente(n)#gerando valores ordenados ordem  descrescente
    inicio2 = timeit.default_timer()
    heapSort(vet2)
    arr2.append(vet2)
    fim2 = timeit.default_timer()
    tempo2.append(fim2-inicio2)

    
x=np.arange(0,10000,intervalo)
plt.plot(x,tempo,'b')
plt.plot(x,tempo1,'g')
plt.plot(x,tempo2,'r')
plt.show()
```

# a versao do quick sort escolhida é do pivo igual ao ultimo elemento é essa versao que vamos fazer as comparaçoes.

```python
def partition(arr,low,high,rand=False):
    i = ( low-1 )
    if(rand==False):
        pivot = arr[high]
    else:
        pivot=randint(1,high)
    for j in range(low , high):
        if arr[j] <= pivot:
            i = i+1
            arr[i],arr[j] = arr[j],arr[i]
    arr[i+1],arr[high] = arr[high],arr[i+1]
    return ( i+1 )

def quickSort(arr,low,high,pivot=False):
	if low < high:
		pi = partition(arr,low,high,pivot)
		quickSort(arr, low, pi-1)
		quickSort(arr, pi+1, high)
arr=[]
arr1=[]
arr2=[]
tempo=[]
tempo1=[]
tempo2=[]
intervalo=10
arrq=[]
tempoq=[]
for n in range(0,100,intervalo):#gerando valores aleatorios
    vet=gera_valor_aleatorio(n)      
    inicio = timeit.default_timer()     
    heapSort(vet)
    arr.append(vet)
    fim = timeit.default_timer()
    tempo.append(fim-inicio)
for n in range(0,100,intervalo):#gerando valores aleatorios
    vetq=gera_valor_aleatorio(n)      
    inicioq = timeit.default_timer()     
    quickSort(vetq,0,n-1)
    arrq.append(vet)
    fimq = timeit.default_timer()
    tempoq.append(fimq-inicioq)
x=np.arange(0,100,intervalo)


plt.plot(x,tempo,'r')
plt.plot(x,tempoq,'b')
plt.title('grafico 100 valores aleatorio')
plt.legend(['heap sorte','quick sorte'])
plt.show()
```

a comparação grafica entre  heap sorte  e quick sorte, como plotado no grafico para  cem valores aleatorios o quick sort tem uma vantagem consideravel em relação o heap sorte.
