```python
import matplotlib.pyplot as plt
import numpy as np
import time
import timeit
import math
```

```python
def heapify(arr, n, i): 
    largest = i 
    l = 2 * i + 1     
    r = 2 * i + 2     
  
    if l < n and arr[i] < arr[l]: 
        largest = l 
  
    if r < n and arr[largest] < arr[r]: 
        largest = r 
  

    if largest != i: 
        arr[i],arr[largest] = arr[largest],arr[i] # swap 
  
        heapify(arr, n, largest) 
  
def heapSort(arr): 
    n = len(arr) 
    # maxheap. 
    for i in range(n, -1, -1): 
        heapify(arr, n, i) 
  
    for i in range(n-1, 0, -1): 
        arr[i], arr[0] = arr[0], arr[i] # swap 
        heapify(arr, i, 0)
```

```python
def partition(arr,low,high,rand=False):
    i = ( low-1 )
    if(rand==False):
        pivot = arr[high]
    else:
        pivot=randint(1,high)
    for j in range(low , high):
        if arr[j] <= pivot:
            i = i+1
            arr[i],arr[j] = arr[j],arr[i]
    arr[i+1],arr[high] = arr[high],arr[i+1]
    return ( i+1 )

def quickSort(arr,low,high,pivot=False):
	if low < high:
		pi = partition(arr,low,high,pivot)
		quickSort(arr, low, pi-1)
		quickSort(arr, pi+1, high)
```

```python
def gera_valor_aleatorio(n):
    arr=(np.random.randint(0,1000,size=n))
    return  (arr)
def gera_valor_ordenado(n):
    return(np.arange(n))
def gera_valor_ordenado_descrecente(n):
    return(np.arange(n,0,-1))
```

```python
def tempo_valores_aleatorio(intervalo,n):
    tempo=[]
    for n in range(0,n,intervalo):#gerando valores aleatorios
        vet=gera_valor_aleatorio(n)      
        inicio = timeit.default_timer()     
        heapSort(vet)
        fim = timeit.default_timer()
        tempo.append(fim-inicio)
    return(tempo)
    
def tempo_valores_ordem_crescente(intervalo,n):
    tempo1=[]
    for n in range(0,n,intervalo):#gerando valores ordenados ordem crescente
        vet1=gera_valor_ordenado(n)
        inicio1 = timeit.default_timer()
        heapSort(vet1)
        fim1 = timeit.default_timer()
        tempo1.append(fim1-inicio1)
    return(tempo1)

    
def tempo_valores_ordem_decrescente(intervalo,n):
    tempo2=[]
    for n in range(0,n,intervalo):
        vet2=gera_valor_ordenado_descrecente(n)#gerando valores ordenados ordem  descrescente
        inicio2 = timeit.default_timer()
        heapSort(vet2)
        fim2 = timeit.default_timer()
        tempo2.append(fim2-inicio2)
    return(tempo2)
```

```python
def tempo_valores_aleatorio_q(intervalo,n):
    tempo=[]
    for n in range(0,n,intervalo):#gerando valores aleatorios
        vet=gera_valor_aleatorio(n)      
        inicio = timeit.default_timer()     
        quickSort(vet,0,n-1)
        fim = timeit.default_timer()
        tempo.append(fim-inicio)
    return(tempo)
    
def tempo_valores_ordem_crescente_q(intervalo,n):
    tempo1=[]
    for n in range(0,n,intervalo):#gerando valores ordenados ordem crescente
        vet1=gera_valor_ordenado(n)
        inicio1 = timeit.default_timer()
        quickSort(vet1,0,n-1)
        fim1 = timeit.default_timer()
        tempo1.append(fim1-inicio1)
    return(tempo1)

    
def tempo_valores_ordem_decrescente_q(intervalo,n):
    tempo2=[]
    for n in range(0,n,intervalo):
        vet2=gera_valor_ordenado_descrecente(n)#gerando valores ordenados ordem  descrescente
        inicio2 = timeit.default_timer()
        quickSort(vet2,0,n-1)
        fim2 = timeit.default_timer()
        tempo2.append(fim2-inicio2)
    return(tempo2)
```

b)Qual e a complexidade no tempo do algoritmo? Explique.

O tempo de execução do ''PENEIRA''  é Ø(1) para corrigir o vetor , mais o tempo que o algoritmo executa recursivamente se comporta como uma arvore binaria.As sub-arvores de cada nó têm tamanho maximo de 2n=3  o pior caso acontece quando o último nível da árvore está exatamente metade cheio é o tempo de execução pode ser expresso pela recorrência:
T(n) <= T(2n=3) + Ø(1).
se resolvemos essa recorrencia vamos achar o tempo de execução do algoritmo peneira é O(log2(n)) e o algoritmo max_heap pode se limitado por O(n),como o maior peso do algoritmo heap sort está no procedimento max_heap e peneira seu O(max_heap*peneira)=O(nlog2(n))


os  graficos mudam a cada compilaçao dos dado,pos a velociade de processamento pode varia de maquina para maquina,alem do sistema de fatiamaneto do processado interferir no cáuculo do tempo .
