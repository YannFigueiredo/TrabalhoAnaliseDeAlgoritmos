# Couting Sort

```python
def couting_sort(A,B,n,k):
    C = []
    for i in range(0,k+1):
        C.append(0)
    for j in range(0,n):
        C[A[j]] = C[A[j]] + 1
    for i in range(1,k+1):
        C[i] = C[i] + C[i - 1]
    for j in range(n-1,-1,-1):
        B[C[A[j]]-1] = A[j]
        C[A[j]] = C[A[j]] - 1
```

```python
#Teste de entradas até 1000 valores
A = [2,2,3,1,0,3,5,2,4,0]
tempo = []
B = len(A)*[0]
for y in range(0,100):#100 execuções com entradas aumentando gradativamente em 10 por execução
    k = max(A)
    inicio = timeit.default_timer()
    couting_sort(A,B,len(A),k)
    final = timeit.default_timer() - inicio
    tempo.append(final)
    print('Entrada com {} valores: {} ms.'.format(len(A),final))
    for x in range(0,10):#Aumento por execução
        num = random.randint(0,10)
        A.append(num)
    B = len(A)*[0]
x = arange(100)
plt.plot(x,tempo)
plt.xlabel("Execuções(+10 por execução)")
plt.ylabel("Tempo(ms)")
plt.grid(True)
plt.show()
```

# Bucket Sort

```python
def bucket_sort(lista):
    largest = max(lista)
    length = len(lista)
    size = largest/length
 
    buckets = [[] for _ in range(length)]
    for i in range(length):
        j = int(lista[i]/size)
        if j != length:
            buckets[j].append(lista[i])
        else:
            buckets[length - 1].append(lista[i])
 
    for i in range(length):
        insertion_sort(buckets[i])
 
    result = []
    for i in range(length):
        result = result + buckets[i]
 
    return result
def insertion_sort(lista):
    for i in range(1, len(lista)):
        temp = lista[i]
        j = i - 1
        while (j >= 0 and temp < lista[j]):
            lista[j + 1] = lista[j]
            j = j - 1
        lista[j + 1] = temp
```
