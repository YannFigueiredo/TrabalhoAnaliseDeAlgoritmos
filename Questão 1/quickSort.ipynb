```python
from random import randint

def inPlaceQuickSort(A,start,end):
    if start<end:
        pivot=randint(start,end)
        temp=A[end]
        A[end]=A[pivot]
        A[pivot]=temp
        
        p=inPlacePartition(A,start,end)
        inPlaceQuickSort(A,start,p-1)
        inPlaceQuickSort(A,p+1,end)


def inPlacePartition(A,start,end):
    pivot=randint(start,end)
    temp=A[end]
    A[end]=A[pivot]
    A[pivot]=temp
    newPivotIndex=start-1
    for index in range(start,end):
        if A[index]<A[end]:#check if current val is less than pivot value
            newPivotIndex=newPivotIndex+1
            temp=A[newPivotIndex]
            A[newPivotIndex]=A[index]
            A[index]=temp
    temp=A[newPivotIndex+1]
    A[newPivotIndex+1]=A[end]
    A[end]=temp
    return newPivotIndex+1
```

```python
import random
import numpy as np
import timeit
import matplotlib.pyplot as plt
from random import *
def partition(arr,low,high):
    i = ( low-1 )
    pivot = arr[high]
    for j in range(low , high):
        if arr[j] <= pivot:
            i = i+1
            arr[i],arr[j] = arr[j],arr[i]
    arr[i+1],arr[high] = arr[high],arr[i+1]
    return ( i+1 )
def quickSort(arr,low,high):
    if low < high:
        pi = partition(arr,low,high)
        quickSort(arr, low, pi-1)
        quickSort(arr, pi+1, high)

def createRandom(n,tam=1000):
    return  np.random.randint(1,tam,size=n)

def executionTime(sequence):
    tempo_inicial = timeit.default_timer()
    quickSort(sequence,low=0,high=len(sequence)-1)
    v=timeit.default_timer()-tempo_inicial
    return v

def randomSequence(amount,start=1,step=10):
    resultado=[]
    for i in range(start,amount,step):
        sequence=createRandom(i-1)
        resultado.append(executionTime(sequence))
    return (resultado)


def createSequenciaOrden(amount,start=1,step=10,descrecent=False):
    resultado=[]
    for i in range(start,amount+1,step):
        if(descrecent==False):
            sequence=np.arange(1,i+1)
        else:
            sequence=np.arange(i,0,-1)
        resultado.append(executionTime(sequence))
    return (resultado)

```
